```python
all_times = np.array([bp_times, cbp_times, linear_times, full_natimp_times, 
                      _50_natimp_times, _10_natimp_times, cbp_random_times,
                      taylor_natimp_times, mean_update_times])
all_times = all_times / all_times.max()
labels = ['bp relu', 'cbp relu', 'bp linear', 'natimp relu (1.0)',
          'natimp relu (0.5)', 'natimp relu (0.1)', 'natimp relu (0.0)',
          'natimp taylor relu', 'mean_update relu']
colors = ['C3', 'C5', 'C4', 'C8', 'C6', 'C7', 'C9', 'C10', 'C11']
fig, ax = plt.subplots()
ax.bar(labels, all_times, color=colors, alpha=0.7)
ax.set_ylabel('Relative Speed (it/s)')
ax.set_xticklabels(labels, rotation=45, ha='right')  
ax.axhline(all_times[1], color=colors[1], linestyle='--', alpha=0.5)
ax.axhline(all_times[5], color=colors[5], linestyle='--', alpha=0.5)
ax.axhline(all_times[7], color=colors[7], linestyle='--', alpha=0.5)
ax.axhline(all_times[8], color=colors[8], linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('runtimes.png', bbox_inches='tight', dpi=500)
```
